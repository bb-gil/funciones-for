```python
# ejercicio 15

texto = "Es mejor conquistarte a ti mismo que ganar mil batallas. Entonces la victoria sera tuya"

# Inicializar contadores para cada vocal en minúscula y mayúscula
contadores = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0, 'A': 0, 'E': 0, 'I': 0, 'O': 0, 'U': 0}

# Iterar sobre cada carácter en el texto
for caracter in texto:
    if caracter in contadores:
        contadores[caracter] += 1

# Imprimir el resultado
for vocal, conteo in contadores.items():
    print(f"{vocal}: {conteo}")
```

1.Inicio del bucle: El bucle comienza con un inicio.

2.Declaración del bucle: Se utiliza un rectángulo con la palabra "For" o
"Para", seguida de la especificación de la variable de iteración y el rango o secuencia sobre el cual iterar.

3.Acciones dentro del bucle: Se colocan las acciones o instrucciones que se ejecutan en cada iteración del bucle.

4.Fin del bucle: Se indica el final del bucle.

```python
# ejercicio 16

# Inicializar contadores
num_pares = 0
num_impares = 0

# Leer 20 números enteros
for _ in range(20):
    numero = int(input("Ingrese un número entero: "))
    if numero % 2 == 0:
        num_pares += 1
    else:
        num_impares += 1

# Imprimir resultados
print("Cantidad de números pares:", num_pares)
print("Cantidad de números impares:", num_impares)
```

```python
# ejercico 17

# Inicializar contadores
multiplos_de_7 = 0
multiplos_de_9 = 0

# Iterar sobre los números entre 1000 y 5000
for numero in range(1000, 5001):
    # Verificar si es múltiplo de 7
    if numero % 7 == 0:
        multiplos_de_7 += 1
    # Verificar si es múltiplo de 9
    if numero % 9 == 0:
        multiplos_de_9 += 1

# Imprimir resultados
print("Cantidad de múltiplos de 7:", multiplos_de_7)
print("Cantidad de múltiplos de 9:", multiplos_de_9)
```

```python
# ejercicio 18

import random
N = int(input("Ingrese el número de dados a lanzar: "))

frecuencia = {i: 0 for i in range(1, 7)}
[frecuencia.update({cara: frecuencia.get(cara, 0) + 1}) for _ in range(N) if (cara := random.randint(1, 6))]
[print(f"Cara {cara}: {frecuencia[cara]} {'*' * frecuencia[cara]}") for cara in frecuencia]
```

```python
# ejercicio 19

numero = int(input("Ingrese un número entero para calcular su factorial: "))

factorial = 1

for i in range(1, numero + 1):
    factorial *= i

print(f"El factorial de {numero} es: {factorial}")
```
